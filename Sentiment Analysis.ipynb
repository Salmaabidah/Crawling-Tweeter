```python
def loadTweets(file='Tweets.json'):
    f=open(file,encoding='utf-8', errors ='ignore', mode='r')
    T=f.readlines();f.close()
    for i,t in enumerate(T):
        T[i] = json.loads(t.strip())
    return T
```

```python
Data_Tweet = loadTweets(file=fileName)
```

# Analisis Sentimen

```python
!wget -P data/ https://raw.githubusercontent.com/Salmaabidah/CrawlingTweeter/main/data_berlabel.csv
df = pd.read_csv('data/data_berlabel.csv', sep = ';')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

seed = 99 # Biasakan menggunakan ini
testSize = 0.2

x_train, x_test, y_train, y_test = train_test_split(df["full_text"], df["Sentimen"],
                                                    test_size=testSize, random_state = seed)
print(x_train.shape, x_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vsm = TfidfVectorizer(lowercase=True, smooth_idf= True, sublinear_tf=True,
                                   ngram_range=(1, 2), max_df=0.90, min_df=2)

x_train = vsm.fit_transform(x_train) # "Fit_Transform"
x_test = vsm.transform(x_test) # Perhatikan disini hanya "Transform"

print(x_train.shape, x_test.shape) # Jumlah kolom Sama ==> ini penting
```

```python
from sklearn import neighbors
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report


model = neighbors.KNeighborsClassifier()
kNN = model.fit(x_train, y_train)
y_kNN = kNN.predict(x_test)

print('Akurasi = ', accuracy_score(y_test, y_kNN))
print(confusion_matrix(y_test, y_kNN))
print(classification_report(y_test, y_kNN))
```

- Precision: Merupakan ukuran dari sejauh mana hasil prediksi positif yang dilakukan oleh model KNN adalah benar. Untuk kelas negatif 50%, kelas netral 62% dan kelas positif 44% dari prediksi positif yang dilakukan oleh model benar.

- Recall: Merupakan ukuran dari sejauh mana model KNN mampu mendeteksi atau mengambil sampel yang benar dari setiap kelas. Untuk kelas negatif model berhasil mendeteksi 56%, kelas netral 42% dan kelas positif 57% dari keseluruhan sampel yang benar-benar positif.

- F1-score: Merupakan penggabungan rata-rata harmonis dari precision dan recall. F1-score adalah ukuran yang baik untuk memperhatikan keseimbangan antara precision dan recall. Untuk kelas negatif, f1-score adalah 0.53, untuk kelas netral f1-score adalah 0.50, dan untuk kelas positif f1-score adalah 0.50.

- Accuracy: Merupakan ukuran dari seberapa akurat model KNN dalam memprediksi seluruh dataset. Accuracy adalah 0.51, yang berarti model memiliki tingkat keakuratan sebesar 51% dalam memprediksi seluruh sampel yang ada.

Hasil analisis sentimen menggunakan metode KNN ini memiliki tingkat akurasi yang relatif rendah, dengan performa yang tidak cukup baik dalam memprediksi dengan tepat setiap kategori sentimen. Dalam rangka meningkatkan performa model, mungkin perlu dilakukan optimasi parameter seperti gridcv atau cross validasi. Mengingat bahwa dataset yang digunakan terbilang sedikit maka gridcv kurang cocok untuk melakukan optimasi parameter

```python
# cross-validasi
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, x_train, y_train, cv=5)

# Melatih model menggunakan data training
model.fit(x_train, y_train)

# Memprediksi kategori sentimen pada data testing
y_pred = model.predict(x_test)

# Menampilkan hasil cross-validation
print("Akurasi CV: {:.2f}".format(scores.mean()))

# Menampilkan hasil prediksi pada data testing
print("Hasil prediksi: ", y_pred)
```

Hasil akurasi dari cross-validasi menurun dari parameter default, yang berarti terdapat masalah-masalah tertentu seperti overfitting, ukuran dataset yang tertalu kecil atau variabilitas data. Dengan demikian untuk studi kasus ini akurasi paling optimal yaitu 51%
